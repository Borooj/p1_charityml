```python
#Read in the dataset and take a look at the top few rows here:
df = pd.read_csv('tmdb-movies.csv')
df.head()
```

### data assessing and cleaning:

```python
#Find the number of rows and columns in the dataset.
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#drop unwanted columns: 
df = df.drop(['id','imdb_id','cast','homepage','director','tagline','keywords','overview','production_companies','release_year','budget_adj','revenue_adj'] , axis=1)
```

```python
#correct the wrong data type to a propper type

df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
#How many missing values in the dataset? 
df.isnull().sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#how many diplicated rows?
df.duplicated().sum()
```

```python
#drop diplicated rows
df.drop_duplicates(inplace=True)
```

```python
#splitting genres column:
df['genres']=df['genres'].str.split('|', 1,expand=True)
```
